```python

from pyspark.sql.functions import col, current_timestamp
```

```python
dbutils.widgets.dropdown(
    "trigger_type",
    "availableNow",
    ["availableNow", "once", "processingTime"]
)

trigger_type = dbutils.widgets.get("trigger_type")

print(f"Using trigger type: {trigger_type}")
```

```python
df = (
    spark.readStream
    .format("cloudFiles")
    .option("cloudFiles.format", "csv")
    .option("header", "true")
    .option("cloudFiles.inferColumnTypes", "true")
    .option(
        "cloudFiles.schemaLocation",
        ORDERS_SCHEMA_LOCATION
    )
    .option(
        "cloudFiles.schemaEvolutionMode",
        "addNewColumns"
    )
    .load(
        LANDING_ORDERS_PATH
    )
)
```

```python
df_enriched = (
    df
    .withColumn(
        "_ingest_timestamp",
        current_timestamp()
    )
    .withColumn(
        "_source_file",
        col("_metadata.file_name")
    )
    .withColumn(
        "_source_path",
        col("_metadata.file_path")
    )
)
```

```python
trigger_options = {
    "availableNow": {
        "availableNow": True
    },
    "once": {
        "once": True
    },
    "processingTime": {
        "processingTime": "10 seconds"
    }
}
```

```python
query = (
    df_enriched.writeStream
    .format("delta")
    .option(
        "checkpointLocation",
        ORDERS_CHECKPOINT_LOCATION
    )
    .option(
        "mergeSchema",
        "true"
    )
    .trigger(
        **trigger_options[trigger_type]
    )
    .toTable(
        BRONZE_ORDERS_TABLE
    )
)


# COMMAND ----------

query.awaitTermination()

print(
    f"Ingestion complete into {BRONZE_ORDERS_TABLE}"
)




print("Micro-batch progress:")

for batch in query.recentProgress:

    source = batch["sources"][0]

    print(
        f"batchId={batch['batchId']} "
        f"rows={source['numInputRows']} "
        f"durationMs={batch['durationMs']}"
    )


```
