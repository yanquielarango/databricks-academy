```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("retail_streaming_platform") \
    .getOrCreate()

data = [
    (1, "customer_1", "Poland"),
    (2, "customer_2", "Spain"),
    (3, "customer_3", "Germany")
]

df = spark.createDataFrame(
    data,
    ["id", "customer", "country"]
)

df.show()
```
